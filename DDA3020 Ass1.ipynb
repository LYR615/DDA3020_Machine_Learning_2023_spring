```python
#Step 5: Randomly split the data into two parts, one contains 80% of the samples and the other
#contains 20% of the samples. Use the first part as training data and train a linear regression model
#and make prediction on the second part with gradient descent methods. X should be the attributes
#you select in previous steps.
#Report the training error and testing error in terms of RMSE. Plot the loss curves in the training
#process. Notice: you need to write the codes of learning the parameters by yourself. Do not use the
#regression packages of Sklearn.
```

```python
X = df2.drop("medv", axis=1)
y = df2["medv"]
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
train_X 
```

```python
train_X = np.insert(np.array(train_X) , 0, 1, axis = 1)    
test_X = np.insert(np.array(test_X) , 0, 1, axis = 1)  

# initialize fitting parameters
w = np.zeros(6) 
```

```python
def computeCost(X, y, w):  
    m = y.size    
    J = 0  
    J = math.sqrt(((1 / m) * np.sum(np.square(X.dot(w) - y))))
    return J          
```

```python
def gradientDescent(X, y, w, alpha, num_iters): 
    m = y.size    
    w = w.copy()          
    J_history = []     
    
    for i in range(num_iters):
        w -= alpha * np.sum(np.expand_dims(X.dot(w) - y, axis= -1) * X, axis=0)
        J_history.append(computeCost(X, y, w))
        
    return w, J_history     
```

```python
def result(alpha,iterations, train_X,train_y, test_X, test_y, w):
    w, J_history = gradientDescent(train_X,train_y, w, alpha, iterations)
    print('w found by gradient descent: {:.4f}, {:.4f}, {:.4f}, {:.4f}, {:.4f}, {:.4f}'.format(*w))
    plt.plot(J_history)     
    plt.xlabel('Iterations')    
    plt.ylabel('Cost_J')            
    plt.show()            

    training_error = computeCost(train_X, train_y, w)  
    print("The training error is: ", training_error) 
  
    testing_error = computeCost(test_X, test_y, w)    
    print("The testing error is: ", testing_error) 
```

```python
# some gradient descent settings
iterations = 150         
alpha = 0.001     
      
result(alpha,iterations, train_X,train_y, test_X, test_y, w)
```
